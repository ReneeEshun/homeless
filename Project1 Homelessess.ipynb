```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
xls=pd.ExcelFile("Homelessdata.xlsx")
homeless_2017=pd.read_excel(xls,"2017")
homeless_2017.head()
```

```python
#Read in year sheets at once.
numbers=range(2007,2018)
years=[str(number) for number in numbers]
homeless_data=pd.read_excel(xls,years)
homeless_data
```

```python
#Grab data for each year for the 4 counties
counties=[
    "Redding/Shasta, Siskiyou, Lassen, Plumas, Del Norte, Modoc, Sierra Counties CoC",
     "San Diego City and County CoC",
     "Oakland/Alameda County CoC",
      "Fresno/Madera County CoC"]
results=[]
for year in years:
    for county in counties:
        year_df=homeless_data[year]
        column_name="Total Homeless, "+year
        homeless=year_df[year_df["CoC Name"]==county][column_name].values[0]
        results.append({"county":county,
                        "year":year,
                        "homeless":homeless})
        
homeless_df=pd.DataFrame(results)
homeless_df
```

```python
pivot_homeless=homeless_df.pivot_table("homeless",["county"],"year")
pivot_homeless.head()
```

```python
plt.plot(pivot_homeless.columns, pivot_homeless.iloc[[0]].transpose(),label="Fresno")
plt.plot(pivot_homeless.columns, pivot_homeless.iloc[[1]].transpose(),label="Alameda County")
plt.plot(pivot_homeless.columns, pivot_homeless.iloc[[2]].transpose(),label="Shasta")
plt.plot(pivot_homeless.columns, pivot_homeless.iloc[[3]].transpose(),label="San Diego")
plt.xlabel("Years 2007-2017")
plt.ylabel("Total Homeless Population")
plt.title("Total Homelessness in County")
plt.legend(bbox_to_anchor=(1,1))
plt.savefig("LineChartHomelessness.png")
```

```python
total_homeless=["CoC Name", "Total Homeless, 2017", "Sheltered Homeless, 2017", "Unsheltered Homeless, 2017"]
types_homeless=homeless_2017[total_homeless]
types_homeless=types_homeless[types_homeless["CoC Name"]!="Total"]
types_homeless
```

```python
types_homeless.dropna(inplace=True)
```

```python
tp = ['Fresno/Madera County CoC', 'Oakland/Alameda County CoC', 'Redding/Shasta, Siskiyou, Lassen, Plumas, Del Norte, Modoc, Sierra Counties CoC', 'San Diego City and County CoC']
types_homeless=types_homeless.loc[types_homeless['CoC Name'].isin(tp)]

types_homeless
```

```python
ind=np.arange(len(types_homeless))
sheltered_homeless=types_homeless["Sheltered Homeless, 2017"]
unsheltered_homeless=types_homeless["Unsheltered Homeless, 2017"]

print(sheltered_homeless)
print(unsheltered_homeless)
width=.50

plot1=plt.bar(ind, sheltered_homeless,width,label="Sheltered")
plot2=plt.bar(ind,unsheltered_homeless, width,bottom=sheltered_homeless,label="Unsheltered")
plt.xlabel("Counties")
plt.ylabel("Homeless Population")
plt.title("Sheltered vs Unsheltered Homelessness in County 2017")
plt.legend(bbox_to_anchor=(1,1))
plt.savefig("stackcolumnshomelessness.png")

labels =["Fresno","Alameda", "Shasta", "San Diego"]

plt.xticks(ind, labels, rotation="vertical" )
plt.show()
```
